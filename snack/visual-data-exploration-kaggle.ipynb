```python
"""
To visually explore the data before analyszing
"""
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

filename = '../input/Iris.csv'

df       = pd.read_csv(filename,sep=',')

# Attribute Information:
# 5. class: 'Species'
#    0 - Iris Setosa
#    1 - Iris Versicolour
#    2 - Iris Virginica
```

```python
# Scatter Matrix

plt.figure( 1 )

# Map label to integer
labelcolors  = pd.factorize(df.Species)[0]

multi = pd.plotting.scatter_matrix(df, c=labelcolors, figsize=(15, 15), marker='o',
                            hist_kwds={'bins': 20}, s=60, alpha=.8)
```

```python
# Parallel Co-ordinates
fignum = plt.gcf().number + 1
plt.figure ( fignum ) 

pd.plotting.parallel_coordinates( df.drop(['Id'], axis=1),'Species')
```

```python
# Andrews Curves
pd.plotting.andrews_curves( df.drop(['Id'], axis=1),'Species')
```
